###Airfoil
The airfoil used to develop Figure 3 from Palmer & Dyke (2012) is believed to be Faired 7%. This airfoil was provided by the author in a .wpa file which can be read by xflr5. xflr5 is an open source analysis tool for airfoils and 3D wings designed in the program using designated airfoils. Lifting Line Theory, Vortex Lattice Method, and a 3D panel Method are analysis options within xflr5. There will be a brief outline of these three methods later.
First, let's read in our airfoil, Faired 7%.

```python
class Readairfoil:
    def __init__(self,filename):
        self.filename = filename
        self.readcoordinates()
        
    def readcoordinates(self):
        coords = np.loadtxt(self.filename)
        self.x = coords[:,0]
        self.y = coords[:,1]
        self.N = len(self.x)
                
    def plot(self,figNum,color,style,width):
        plt.figure(num=figNum)
        plt.grid(True)
        plt.xlabel('x',fontsize=12)
        plt.ylabel('y',fontsize=12)
        plt.plot(self.x,self.y,c=color,ls=style,lw=width)
        plt.axis('equal')
        plt.xlim(0,1)
```

###Wing Geometries
The wing geometires used in this notebook as an attempt to recreate Figure 3 were defined in a spreadsheet provided by the authors. This spreadsheet was delivered in a second wave of requested data. The orignial wing geometries provided did not align with any of the plots in the paper. 

The 800 mm sweep point included in Figure 3 was not defined in the spreadsheet provided. I used the forumlas provided for the other points to be able to include this point in my graph.

The wing geometries altered only in sweep (or offset). There is no change in chord length for the 'bennett' wing geometries. The airfoil above shown above is the only airfoil included in the various bennett wing geometries. They are all defined by utilizing ten of the airfoils spaced in the same locations along the y axis within the program, varying only in the x location.


```python
class section:
    def __init__(self,foil,xle,yle,zle,chord):
        self.xle = xle
        self.yle = yle
        self.zle = zle
        self.chord = chord
        self.getcoordinates(foil)
        self.xte = max(self.x)
        self.yte = self.y[self.x.argmax(axis=0)]
        self.zte = self.yle*np.ones(len(self.x))
    
    def getcoordinates(self,f):
        self.x = self.xle + self.chord*(+f.x)
        self.z = self.yle + self.chord*(+f.y)
        self.y = self.yle*np.ones(len(self.x))
```

```python
class Topview:
    def __init__(self,topview):
        self.x = 'y'
        self.y = 'x'

class Wing:
    def __init__(self,name):
        self.name = name
        self.sections = []
        self.method = {}
        self.alpha = 0.
    
    def append(self,section):
        self.sections.append(section)
    
    def plot(self,fignum,color,style,width):
        plt.figure(num=fignum,figsize= (12,10))
        plt.grid(True)
        plt.xlabel('y',fontsize=12)
        plt.ylabel('x',fontsize=12)
        plt.axis('equal')
        for s in self.sections:
            plt.plot(s.y,s.x,c=color,ls=style)
        for i in range(len(self.sections)-1):
            plt.plot([self.sections[i].yle,self.sections[i+1].yle],\
                         [self.sections[i].xle,self.sections[i+1].xle],\
                         c=color,lw=width,ls=style)
            plt.plot([self.sections[i].yte,self.sections[i+1].yte],\
                         [self.sections[i].xte,self.sections[i+1].xte],\
                         c=color,lw=width,ls=style)
        plt.ylim(1.5,-1.5)
        print color+style+'('+str(width)+')',':',self.name
    
    def addresults(self,meth,filename):
        data = np.loadtxt(filename,skiprows=8)
        self.method[meth]=np.array\
        ([data[:,0],data[:,1],data[:,4],data[:,11]])
        #[:,0]-alpha,[:,1]-CL,[:,4]-CD,[:,11]-XCP
```

### Analysis Methods
The analysis method indicated in the paper was vortex lattice method. xflr5 has two options for vortex lattice methods. Horseshoe vortex (VLM1)(no sideslip) and Ring Vortex (VLM2) are both options within the analysis definition in xflr5. 3D panels is also a selection. The paper cites that vortex lattice method was used, but did not specify whic one. I decided to run both of these options in an attempt to replicate the results. I also used the 3D Panel method in xflr5 because we developed a basic 2D Panel method code in class and I was interested in how it would compare to vortex lattice method in this context. In all cases, the surface is meshed into a number of panels. xflr5 has both viscous and inviscid options.

####Horseshoe Vortex Lattice Method (VLM1)
Vortex Lattice method is a continuation of the lifting-line theory introduced by Prandtl. For horeshoe VLM, each lifting surface is represented by a single row of horeshoe shaped vortices as shown below. I found a quick review on vortex theory to be very helpful in trying to understanding the different method options.
The analysis using either VLM1 or VLM2 is calculated on the mean camber line. Wake is represented by utilizing the legs of the horseshoe vortices.



```python
VLM1=Image(filename='vlm1.png')
VLM1

#Image from Non-Linear Lifting-Line Solver S.J. Hulshoff
#https://aerodynamics.lr.tudelft.nl/~shulshoff/nlll/doc/manual.html
```

#### Ring Vortex Lattice Method (VLM2)
xflr5 has two vortex lattice methods to choose from with in the program. The second option is ring vortex lattice metod (VLM2). This method is also named quad vortex. In this method, the interior vortices are all rings and only the trailing vorticies remain represented as horeshoe vortices. For most test cases, there will be little to no difference between VLM1 and VLM2. 

```python
VLM2=Image(filename='vlm2.png')
VLM2
```

####3D Panel Method
The final method I experimented with in xflr5 was the 3D Panel Method Option. After developing a basic 2D panel method code in class, I was interested in experimenting with a code that has 3D capability. Unlike vortex lattice methods, thinkness is considered in the calculation. Without having the semi-infinite vortices extending (like in VLM1 and VLM2), the wake is represented as a series of thin panels. This method is also able to distribute Cp over the top and bottom surfaces of a wing.

```python
methods=['VLM1','VLM2','Panel3D']

AoA=1.0  #angle of attack can be changed from -2 to 8 in 0.5 increments
wing['bennettbase'].alpha = AoA
wing['bennett200'].alpha = AoA
wing['bennett400'].alpha = AoA
wing['bennett600'].alpha = AoA
wing['bennett700'].alpha = AoA
wing['bennett800'].alpha = AoA

refSweep=wing['bennettbase'].sections[-1].xle

plt.figure(num=None,figsize=(6,6))
plt.grid(True)
plt.xlabel('Anterior Sweep (mm)',fontsize=12)
plt.ylabel('CP Location (% Wingspan from Glenoid)',fontsize=12)
plt.title('CP Location vs. Anterior Sweep at Angle of Attack =%s'%(AoA))

for m in methods:
    sweep,XCP = [],[]
    for w in wings:
        for i in range(len(w.method[m][0])):
            if(w.method[m][0][i]==w.alpha):I = i
        sweep.append(-1000.*(w.sections[-1].xle-refSweep))
        XCP.append(-100.*w.method[m][3][I]\
                   /(w.sections[-1].yle-w.sections[0].yle)/2)
    plt.plot(sweep,XCP,marker='o',ls='-',lw=1)
plt.plot(sweep,[-2.30,-1.40,-0.05,1.50,2.21,3.0],marker='o',ls='--',lw=1)
methods.append('Palmer & Dyke (2012)')
plt.legend(methods,loc='best',prop={'size':12});
                
```

As you will note, these graphs do not line up. The results for VLM1 and VLM2 are the same while the results for Panel3D are underesitmating the location of the CP. None of the results I obtained match up to the results provided in the paper. I have obtained linear results and have worked with different angles of attach and have noted that AoA = 2.0 aligns the plots, but clearly that my results are linear and the results of the paper are not. 
Changing the angle of attack has a significant impact on the results. It is possible to align any given point with the Palmer & Dyke results by utilizing a different angle of attach.
