# Finite and Infinite Row of Vortices

##Finite Row of Vortices
Let's start with the basics of importing in the appropriate libraries and creating our meshgrid.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
N=50                            # Number of points in each direction
xStart,xEnd = -2.0,2.0          # x direction boundaries
yStart,yEnd = -1.0,1.0          # y direction boundaries
x = np.linspace(xStart,xEnd,N)  # x 1D-array
y = np.linspace(yStart,yEnd,N)  # y 1D-array
X,Y = np.meshgrid(x,y)          # generation of the mesh grid
```

###Define Functions
Let's create functions to calculate our velocity field and stream function.
This is how we calculated velocity fields and stream functions in previous AeroPython notebooks.

```python
def GetVelocityVortex(strength,xv,yv,X,Y):
    u = strength/(2*pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v = - strength/(2*pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    return u,v

def GetStreamFunctionVortex(strength,xv,yv,X,Y):
    psi = strength/(4*pi)*np.log((X-xv)**2+(Y-yv)**2)
    return psi
```

Let's define the vortex strength and some additional parameters. I will define my vortex row to on the x axis.

```python
gamma = 5.0                              # strength of vortex
vortnum = 7                             # number of vortecies
xVortex=np.linspace(xStart,xEnd,vortnum) # x position of vorex line
yVortex=np.zeros(vortnum)                # y position  of vortex line 
```

In order to set-up a for loop to loop through each point we need to define the variable size. This will be done by creating an empty array. We will then loop through point by point.

```python
u = np.zeros((N,N),dtype=float)         # set up for for loop
v = np.zeros((N,N),dtype=float)         # place holder for for loop

for i in range (0,vortnum):             # define for loop
    uVortex,vVortex=GetVelocityVortex(gamma,xVortex[i],yVortex[i],X,Y)
    psiVortex = GetStreamFunctionVortex(gamma,xVortex[i],yVortex[i],X,Y)
    u = u + uVortex
    v = v + vVortex
```

Let's graph it to check out our results (slash check it to make sure we did it correctly)

```python
size = 10
plt.figure(figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.title('Finite Row of Vortices')
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,u,v,density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter(xVortex,yVortex,c='r',s=80,marker='o')

plt.show()
```

## Infinite Row of Vorticies
Hopefully there is still more room for fun because it is time to look at an infinite row of vorticies. 

First, we're going to define some additional variables, and then we will define a new function. We will then plot the results.

```python
a = 0.5                           # distance between vortices
xVortInf = np.arange(xStart,xEnd,a) # x position.. must be updated from orig
yVortInf= np.zeros_like(xVortInf)

def getInfinityVortexVelocity(strength,a,X,Y):
    u = strength/(2*a)*np.sinh(2*pi*Y/a)/(np.cosh(2*pi*Y/a)-np.cos(2*pi*X/a))
    v = -strength/(2*a)*np.sin(2*pi*X/a)/(np.cosh(2*pi*Y/a)-np.cos(2*pi*X/a))
    return u,v
# first we defined the new function, now let's use it

infivortu,infivortv = getInfinityVortexVelocity(gamma,a,X,Y)


#Let's Plot!
size = 10
plt.figure(figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,infivortu,infivortv,density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter(xVortInf,yVortInf,c='r',s=80,marker='o');
```
